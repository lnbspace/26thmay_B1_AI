# Matplotlib

1. plot -> Line Plot
2. box -> Box Plot, that is statstical view of your data in terms of boxes; mean, max,min, 50%, 25%, 75%, outlier
3. boxplot -> Box Plot, that is statstical view of your data in terms of boxes; mean, max,min, 50%, 25%, 75%, outlier
4. bar -> shows the data in terms of bars (Vertical)
5. barh -> Horizontal Bar Graph
6. hist -> Frequency of Data in Bars
7. hexbin -> Scatter plot, but shape of points in hexagon
8. pie -> Pie Chart
9. scatter -> Ordinary Scatter plot (compares two different numerical Data)
10. stem -> skeleton version of bar graph
11. step -> Step Graph
12. violinplot -> Violin version of Boxplot

```python
p = plt.figure(figsize=(12,6))
p = plt.plot(a,b)
p = plt.grid()
p = plt.title('Random Data Graph')
p = plt.xlabel('X Data-->')
p = plt.ylabel('Y Data-->')
p = plt.legend('Random Line')
p = plt.xticks(range(0,41,10),['Zero', 'Ten','Twenty','Thirty','Forty'])

# plt.show()
```

```python
p = plt.figure(figsize=(12,6))
p = plt.plot(a,b, color='r', marker = 'o', markerfacecolor='g',markeredgecolor='blue',lw=3,ls='-.',markersize=10)
p = plt.grid()
p = plt.title('Random Data Graph')
p = plt.xlabel('X Data-->')
p = plt.ylabel('Y Data-->')
p = plt.legend('Random Line')
p = plt.xticks(range(0,41,10),['Zero', 'Ten','Twenty','Thirty','Forty'])
```

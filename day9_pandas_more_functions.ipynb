### Groupby Operations

```python
df = pd.read_csv('Teams.csv')
df
```

```python
df.groupby('Teams')['Points'].mean()
```

```python
df.groupby(['Teams','Year'])['Points'].mean()
```

```python
g = df.groupby(['Teams','Year'])
```

```python
g.groups
```

```python
g.get_group(('Devils',2014))
```

```python
df.groupby('Teams').get_group('Riders')
```

```python
for i,v in df.groupby('Teams'):
    print(i)
    print(v)
```

```python
import numpy as np
```

```python
df.groupby('Teams')['Points'].agg([len, np.mean, np.std, min, max])
```
